```python
import os
from collections import defaultdict

# ==========================
# CONFIGURATION
# ==========================
ROOT_DIR = r"/home/sandeshprasai/Final_Semester_Project/AI_Attendance_System/ai-ml-model/DataSets/raw"   # change this
BIN_SIZE = 100                       # 0-100, 100-200, etc.
IMAGE_EXTENSIONS = (".jpg", ".jpeg", ".png", ".bmp", ".webp")

# ==========================
# COUNT IMAGES PER CLASS
# ==========================
class_image_counts = {}

for class_name in os.listdir(ROOT_DIR):
    class_path = os.path.join(ROOT_DIR, class_name)

    if not os.path.isdir(class_path):
        continue

    image_count = sum(
        1 for f in os.listdir(class_path)
        if f.lower().endswith(IMAGE_EXTENSIONS)
    )

    class_image_counts[class_name] = image_count

# ==========================
# BIN CLASSES BY IMAGE COUNT
# ==========================
bins = defaultdict(int)

for count in class_image_counts.values():
    lower = (count // BIN_SIZE) * BIN_SIZE
    upper = lower + BIN_SIZE
    bin_label = f"{lower}-{upper}"
    bins[bin_label] += 1

# ==========================
# SORT & DISPLAY RESULTS
# ==========================
print("Class Distribution by Image Count\n")

for bin_label in sorted(
    bins.keys(),
    key=lambda x: int(x.split("-")[0])
):
    print(f"{bin_label} images : {bins[bin_label]} classes")


```
