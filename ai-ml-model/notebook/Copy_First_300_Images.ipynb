```python
import shutil
from pathlib import Path
from tqdm import tqdm

def copy_first_300_images():
    SOURCE_DIR = r"D:\Final_Semester_Project\AI_Attendance_System\ai-ml-model\DataSets\raw"
    TARGET_DIR = r"D:\Final_Semester_Project\AI_Attendance_System\ai-ml-model\DataSets\SecondLap\First300"

    IMAGE_EXTENSIONS = {'.jpg', '.jpeg', '.png', '.bmp'}

    source_path = Path(SOURCE_DIR)
    target_path = Path(TARGET_DIR)
    target_path.mkdir(parents=True, exist_ok=True)

    class_dirs = [d for d in source_path.iterdir() if d.is_dir()]

    copied_classes = 0
    skipped_classes = 0
    total_images = 0

    with tqdm(class_dirs, desc="Processing classes") as pbar:
        for class_dir in pbar:
            # Collect images ONCE, no duplicates
            image_files = sorted(
                [f for f in class_dir.iterdir()
                 if f.is_file() and f.suffix.lower() in IMAGE_EXTENSIONS],
                key=lambda x: x.name.lower()
            )

            # Skip if fewer than 300 images
            if len(image_files) < 300:
                skipped_classes += 1
                continue

            # Create target class directory
            target_class_dir = target_path / class_dir.name
            target_class_dir.mkdir(exist_ok=True)

            # Copy exactly first 300 images
            for img in image_files[:300]:
                shutil.copy2(img, target_class_dir / img.name)

            copied_classes += 1
            total_images += 300

    print("\n=== SUMMARY ===")
    print(f"Classes processed : {copied_classes}")
    print(f"Classes skipped   : {skipped_classes}")
    print(f"Images copied    : {total_images}")

if __name__ == "__main__":
    copy_first_300_images()
```
